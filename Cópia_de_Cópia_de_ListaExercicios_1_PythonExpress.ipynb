Elabore um código que peça dois números e imprima a soma.

Elabore um código que peça as 4 notas bimestrais e mostre a média.

```python

nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))
nota4 = float(input("Digite a quarta nota: "))

media = (nota1 + nota2 + nota3 + nota4) / 4

print(f"A média das notas é: {media:.2f}")
```

Elabore um código que peça o raio de um círculo, calcule e mostre sua área.

```python
import math

raio = float(input("Digite o valor do raio do círculo: "))


area = math.pi * (raio ** 2)

print(f"A área do círculo com raio {raio} é: {area:.2f}")
```

Elabore um código que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês em uma mensagem amigável.

```python

valor_hora = float(input("Quanto você ganha por hora? R$ "))
horas_mes = float(input("Quantas horas você trabalhou no mês? "))

salario_total = valor_hora * horas_mes

print(f"\nOlá! Trabalhando {horas_mes} horas a R$ {valor_hora:.2f}/h,")
print(f"seu salário total no mês é: R$ {salario_total:.2f}")
```

Elabore um código que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius. ** Onde C = 5 * ((F-32) / 9)**

```python


fahrenheit = float(input("Digite a temperatura em Fahrenheit: "))

celsius = 5 * ((fahrenheit - 32) / 9)


print(f"{fahrenheit}ºF é igual a {celsius:.2f}ºC")
```

Elabore um código que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.

```python

celsius = float(input("Digite a temperatura em graus Celsius: "))

fahrenheit = (celsius * 1.8) + 32


print(f"{celsius}°C equivale a {fahrenheit:.2f}°F")
```

Elabore um código que peça 2 números inteiros e um número real. Calcule e mostre:

o produto do dobro do primeiro com metade do segundo .
a soma do triplo do primeiro com o terceiro.
o terceiro elevado ao cubo.


```python

num_int1 = int(input("Digite o primeiro número inteiro: "))
num_int2 = int(input("Digite o segundo número inteiro: "))
num_real = float(input("Digite um número real: "))


resultado1 = (2 * num_int1) * (num_int2 / 2)


resultado2 = (3 * num_int1) + num_real


resultado3 = num_real ** 3


print(f"\nResultados:")
print(f"O produto do dobro do primeiro com metade do segundo: {resultado1}")
print(f"A soma do triplo do primeiro com o terceiro: {resultado2}")
print(f"O terceiro elevado ao cubo: {resultado3:.2f}")
```

Elabore um código que receba os dados de entrada a altura de uma pessoa, e informe qual seria o peso ideal, usando a seguinte fórmula: (72.7*altura) - 58

```python


def calcular_peso_ideal():
    try:

        altura = float(input("Digite a altura (ex: 1.75): "))


        peso_ideal = (72.7 * altura) - 58


        print(f"Para a altura de {altura:.2f}m, o peso ideal é: {peso_ideal:.2f} kg")

    except ValueError:
        print("Por favor, digite um número válido para a altura (use ponto para decimais).")

calcular_peso_ideal()
```
