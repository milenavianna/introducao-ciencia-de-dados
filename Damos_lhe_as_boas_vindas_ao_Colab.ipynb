EXERCICIO 18

```
def calcula_desconto(valor, desconto):
    return valor - (valor * desconto / 100)

resultado = calcula_desconto(100, 10)

print(resultado)
```

EXERCICIO 19

```
def classifica_nota(nota):
    if nota >= 7:
        return "Aprovado"
    elif nota >= 5:
        return "Recuperação"
    else:
        return "Reprovado"

print(classifica_nota(6.5))
```

EXERCICIO 20

```
numeros = [10, -5, 0, 7, -2, 0, 3, -8]

positivos = 0
negativos = 0
zeros = 0

for n in numeros:
    if n > 0:
        positivos += 1
    elif n < 0:
        negativos += 1
    else:
        zeros += 1

print("Positivos:", positivos)
print("Negativos:", negativos)
print("Zeros:", zeros)
```
